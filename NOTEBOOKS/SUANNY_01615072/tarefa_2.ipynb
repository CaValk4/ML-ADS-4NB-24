```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

vinhos.head()
```

Análise Exploratória de Variáveis

```python

classe_counts = vinhos['classe'].value_counts()
print(classe_counts)

estatisticas = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()
print(estatisticas)
```

Criação de Boxplots

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(y='magnesio', data=vinhos)
plt.title('Boxplot de Magnésio')

plt.subplot(2, 2, 2)
sns.boxplot(y='prolina', data=vinhos)
plt.title('Boxplot de Prolina')

plt.subplot(2, 2, 3)
sns.boxplot(y='cinzas', data=vinhos)
plt.title('Boxplot de Cinzas')

plt.subplot(2, 2, 4)
sns.boxplot(y='alcool', data=vinhos)
plt.title('Boxplot de Álcool')

plt.tight_layout()
plt.show()
```

Boxplots Segregados por Classe

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Magnesio por Classe')
plt.suptitle('')

plt.subplot(2, 2, 2)
vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Fenóis Totais por Classe')
plt.suptitle('')

plt.subplot(2, 2, 3)
vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Ácido Malico por Classe')
plt.suptitle('')

plt.subplot(2, 2, 4)
vinhos.boxplot(column='alcool', by='classe')
plt.title('Álcool por Classe')
plt.suptitle('')

plt.tight_layout()
plt.show()
```

Criação de Histograma

```python
# Histograma de fenois_totais
plt.figure(figsize=(10, 6))
sns.histplot(vinhos['fenois_totais'], bins=10, kde=True)
plt.title('Distribuição de Fenóis Totais')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()

# Histograma de fenois_totais para a classe 3
plt.figure(figsize=(10, 6))
sns.histplot(vinhos[vinhos['classe'] == '3']['fenois_totais'], bins=10, kde=True)
plt.title('Distribuição de Fenóis Totais (Classe 3)')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()
```
